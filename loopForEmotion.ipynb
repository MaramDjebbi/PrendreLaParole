### Python script to Loop through PEDFE videos and radomly choose 10 vids for each emotion

```python
import pandas as pd
import os
import shutil
import random

# Load CSV
df = pd.read_csv("/home/maram/Downloads/osfstorage-archive/Supplemental_Material_T2.csv")

# Base paths
base_path = "/home/maram/Downloads/osfstorage-archive/PEDFE_Video"
output_base = "/home/maram/Downloads/PrendreLaParole/PEDFE/Backup"  

# Get unique emotions
emotions = df['Emotion'].dropna().unique()

for emotion in emotions:
    # Filter for emotion and type Genuine
    filtered = df[(df['Emotion'] == emotion) & (df['Type'] == "Genuine")]

    # Skip if fewer than 10 videos
    if len(filtered) < 10:
        continue

    # Randomly pick 10
    sampled = filtered.sample(10, random_state=42)

    # Output dir
    emotion_dir = os.path.join(output_base, emotion)
    os.makedirs(emotion_dir, exist_ok=True)

    for _, row in sampled.iterrows():
        subject = str(int(row['Subject']))
        subject_folder = f"{subject} genuine"
        pedfe_code = row['PEDFE_code']
        src_path = os.path.join(base_path, subject, subject_folder, str(pedfe_code) + ".mp4")

        if os.path.exists(src_path):
            dst_path = os.path.join(emotion_dir, str(pedfe_code) + ".mp4")
            shutil.copy(src_path, dst_path)
            print(f"Copied {pedfe_code}.mp4 to {emotion_dir}")
        else:
            print(f"Missing file: {src_path}")
```
